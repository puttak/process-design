```python
def makeplot(filn,SCratio,HEATADDED,ENSOEC,CO_OG,CO_SG,CO2_OG,CO2_SG,OXADDED,CH4_OG,CH4_SG,J,maxCH4):
    #makeplot(filn,SCratio,HEATADDED,ENSOEC,CO_OG,CO_SG,CO2_OG,CO2_SG,OXADDED,CH4_OG,CH4_SG,J)
    plt.figure()
    # plt.style.use('fivethirtyeight')
    #plt.style.use('seaborn-paper')
    plt.style.use('seaborn')
    plt.subplot(221)
    plt.title('')
    # plt.title('100 kmol/h Oxygen to and 1300 C out of Gasifier')
    # plt.title('No Oxygen to and 1300 C out of Gasifier')
    plt.plot(SCratio, HEATADDED, label='Energy to Gasifier')
    plt.plot(SCratio, ENSOEC, label='Energy to SOEC')
    plt.plot(SCratio, HEATADDED+ENSOEC, label='Total')
    plt.grid(True)
    # plt.xlabel('Steam to carbon ratio')
    plt.ylabel('Energy added [MW]')
    plt.axis([0, 2.1, 0, 120])
    plt.legend()
    plt.tight_layout()
    
    plt.subplot(223)
    plt.plot(SCratio,CO_OG, label='CO OG')   # this is CO+H2
    plt.plot(SCratio,CO_SG, label='CO SG')   # this is CO+H2
    plt.plot(SCratio,CO2_OG, label='CO2 OG')
    plt.plot(SCratio,CO2_SG, label='CO2 SG')
    plt.plot(SCratio,OXADDED, label='O2')
    plt.xlabel('Steam to carbon ratio')
    plt.ylabel('Flow [kmol/h]')
    plt.axis([0, 2.1, 0, 1100])
    plt.grid(True)
    plt.legend() 
    plt.tight_layout()
    
    plt.subplot(222)
    plt.plot(SCratio,CH4_OG, label='CH4 OG')   # 
    plt.plot(SCratio,CH4_SG, label='CH4 SG')   # 
    # plt.xlabel('Steam to carbon ratio')
    plt.ylabel('Flow [kmol/h]')
    maxL=int(math.ceil(maxCH4 / 10.0)) * 10
    plt.axis([0, 2.1, 0, maxL])
    plt.grid(True)
    plt.legend() 
    plt.tight_layout()
    
    plt.subplot(224)
    plt.plot(SCratio,J,label='Profit')
    plt.grid(True)
    plt.ylabel('Criterion $J$ [MUSD]')
    plt.xlabel('Steam to carbon ratio')
    plt.axis([0, 2.1, 25, 34])
    # plt.legend()
    plt.tight_layout()
    
    if len(filn)>1:
        plt.plt.savefig(filn,dpi=300)
```

```python
makeplot('E5525bar.png',E30['S/C ratio'],E30['Energy to Gasifier'],E30['E_SOEC'],E30['CO OG'],E30['CO SG'],E30['CO2 OG'],E30['CO2 in gas 2FT'],E30['O2 added'],E30['CH4 OG'],E30['CH4 SG'],E30['Criterion'],maxCH4)
```
